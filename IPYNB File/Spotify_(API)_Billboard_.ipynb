We will remove outliers from tempo, popularity and energy


```python
#For popularity
pop_25=grand_df_songs['Popularity'].quantile(.25)
pop_75=grand_df_songs['Popularity'].quantile(.75)
pop_iqr=pop_75-pop_25

upper_limit_pop = pop_75 + 1.5 * pop_iqr
lower_limit_pop = pop_25 - 1.5 * pop_iqr

grand_df_songs_pop=grand_df_songs[(grand_df_songs['Popularity']<=upper_limit_pop)&(grand_df_songs['Popularity']>=lower_limit_pop)]

#For Energy
ene_25=grand_df_songs_pop['Energy'].quantile(.25)
ene_75=grand_df_songs_pop['Energy'].quantile(.75)
ene_iqr=ene_75-ene_25

upper_limit_ene = ene_75 + 1.5 * ene_iqr
lower_limit_ene = ene_25 - 1.5 * ene_iqr

grand_df_songs_pop_ene=grand_df_songs_pop[(grand_df_songs_pop['Energy']<=upper_limit_ene)&(grand_df_songs_pop['Energy']>=lower_limit_ene)]

#For Tempo
tempo_25=grand_df_songs_pop_ene['Tempo'].quantile(.25)
tempo_75=grand_df_songs_pop_ene['Tempo'].quantile(.75)
tempo_iqr=tempo_75-tempo_25

upper_limit_tempo = tempo_75 + 1.5 * tempo_iqr
lower_limit_tempo = tempo_25 - 1.5 * tempo_iqr

grand_df_songs_pop_ene_tempo=grand_df_songs_pop_ene[(grand_df_songs_pop_ene['Tempo']<=upper_limit_tempo)&(grand_df_songs_pop_ene['Tempo']>=lower_limit_tempo)]

```

Feature Selection

```python
from sklearn.feature_selection import mutual_info_classif
impor= mutual_info_classif(X,y)
feat_imp=pd.Series(impor, X.columns[:len(X.columns)])
plt.title("rank of each feature based on information gain")
feat_imp.plot(kind='barh')
plt.savefig('/content/drive/MyDrive/Spotify_Images/feature selection.png')
```

```python
X_reduced=X.drop(['Liveness', 'Mode', 'Danceability'], axis=1)
X_train, X_test, y_train, y_test=train_test_split(X_reduced, y, test_size=.2)
```

```python
sc=StandardScaler()
X_train[['Tempo',	'Duration']]=sc.fit_transform(X_train[['Tempo',	'Duration']])
X_test[['Tempo',	'Duration']]=sc.fit_transform(X_test[['Tempo',	'Duration']])

rf_model=RandomForestClassifier(random_state=17)
rf_model.fit(X_train, y_train)

y_pred=rf_model.predict(X_test)
accuracy_score(y_pred, y_test)
```

```python
plot_confusion_matrix(rf_model, X_test, y_test)
plt.title("confusion matrix for Random Forest")
plt.show()
plt.savefig("/content/drive/MyDrive/Spotify_Images/confusion mat")
```
